```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv("epa-sea-level.csv")

def draw_plot():
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Data', alpha=0.7)

    slope_all, intercept_all, *_ = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years_extended = pd.Series(range(1880, 2051))
    plt.plot(years_extended, intercept_all + slope_all * years_extended, 'r', label='Fit: 1880–2050')

    df_recent = df[df['Year'] >= 2000]
    slope_recent, intercept_recent, *_ = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    years_future = pd.Series(range(2000, 2051))
    plt.plot(years_future, intercept_recent + slope_recent * years_future, 'g', label='Fit: 2000–2050')

    plt.title("Rise in Sea Level")
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")
    plt.legend()
    plt.grid(True)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
